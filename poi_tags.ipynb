```python
"""读取文件函数"""
def read_file(filename):
    with open(filename,'r') as f:
        for line in f:
            if line:
                yield line
            else: 
                return
```

```python
"""读取文件"""
filename = "data/poi_tags_row.csv"
poi_tags = {}
row_num = 0
for line in read_file(filename):
    if row_num == 0:
        row_num += 1
        continue
    item = line.split('\t')
    poi_name = item[2]
    tag_name = item[5]
    if poi_name not in poi_tags:
        poi_tags[poi_name]=[tag_name]
    else:
        poi_tags[poi_name].append(tag_name)
    
```

```python
"""列表去重"""
def gene_set(myset):
    tags_set=[]
    for item in myset:
        if item in tags_set:
            continue
        else:
            tags_set.append(item)
    return tags_set
```

```python
poi_list = []
tags_list =[]
for key,items in poi_tags.items():
    poi_list.append(key)
    tags_list.extend(items)

poi_set = gene_set(poi_list)
tags_set = gene_set(tags_list)
```

```python
tags_set
```

```python
"""生成字典"""
poi_tags_matrix=np.zeros((len(poi_set),len(tags_set)))
for poi in poi_tags.keys():
    # print("-----------")
    # print(poi)
    PoiInd = poi_set.index(poi)
    for tag in poi_tags[poi]:
        # print(tag)
        TagInd =  tags_set.index(tag)
        poi_tags_matrix[PoiInd,TagInd]=1
    
```

```python
"""共同评论矩阵"""
common_matrix = np.dot(poi_tags_matrix,poi_tags_matrix.T)
```

```python
common_matrix
```

```python
len(common_matrix)
```

```python
# 每个poi的tag数量
rate_num = poi_tags_matrix.sum(axis=1)

JacSim = np.zeros((len(common_matrix),len(common_matrix)))
for i in range(len(common_matrix)):
    for j in range(len(common_matrix)):
        if i==j:
            continue
        JacSim[i,j] = common_matrix[i,j]/(rate_num[i]+rate_num[j]-common_matrix[i,j])
```
